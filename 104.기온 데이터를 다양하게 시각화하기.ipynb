```python
import csv
import matplotlib.pyplot as plt

f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)
result = []

aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(aug, bins=500 , color='red', label='Aug')
plt.hist(jan, bins=500 , color='blue', label='Jan')
plt.legend()
plt.show
```

```python
import csv
import matplotlib.pyplot as plt

f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)
month= [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31) :
    day.append([])
    
for row in data :
    if row[-1] != '':
        if row[0].split('-')[1] == '05':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.boxplot(day, showfliers=False) # outlier 생략
plt.figure(figsize=(10,5), dpi=300)
plt.show()
```
