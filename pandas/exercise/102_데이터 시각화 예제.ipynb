```python
import csv
import matplotlib.pyplot as plt

f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)
high=[]
low=[]

for row in data : 
    if row[-1] != '' and row[-2] != '' :
        if int(row[0].split('-')[0]) >= 1992 :
            if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '27' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.plot(high, 'r')
plt.plot(low, 'b')
plt.show()
    
```
